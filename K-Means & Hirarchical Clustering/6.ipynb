```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, normalize
```

```python
df = pd.read_csv('Hitters.csv')
df
```

```python
df.isna().sum()
```

```python
df['Salary'].fillna(df['Salary'].mean(), inplace= True)
```

```python
df.info()
```

```python
df[['League', 'Division', 'NewLeague']] = df[['League', 'Division', 'NewLeague']].apply(LabelEncoder().fit_transform)
```

```python
X = pd.DataFrame(normalize(df), columns=df.columns)
```

```python
X
```

```python
plt.figure(figsize=(9*3, 6*3))
sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=3)
model.fit(X)
```

```python
X['labels'] = model.labels_
```

```python
plt.scatter(X['CRuns'], np.zeros(X.shape[0]) +5, c = X['labels'])
plt.show()
```
