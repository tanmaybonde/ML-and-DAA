```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = [
    [0.1,0.6],
    [0.15,0.71],
    [0.08,0.9],
    [0.16, 0.85],
    [0.2,0.3],
    [0.25,0.5],
    [0.24,0.1],
    [0.3,0.2]
]
```

```python
centroids = [
    [0.1,0.6],
    [0.3,0.2]
]
```

```python
df = pd.DataFrame(data, columns= ['X', 'Y'])
cent = pd.DataFrame(centroids, columns=['X', 'Y'])
```

```python
model = KMeans(n_clusters=2, max_iter=50, init=centroids, n_init=1)
model.fit(df)
y_ = pd.DataFrame(model.cluster_centers_, columns=['X', 'Y'])
```

```python
fig, axis = plt.subplots(2, 2, figsize = (9*2, 6*2))
axis[0][0].scatter(cent['X'], cent['Y'])
axis[0][0].set_xlim(0, 0.4)
axis[0][0].set_ylim(0, 1)
axis[0][1].scatter(df['X'], df['Y'])
axis[0][1].set_xlim(0, 0.4)
axis[0][1].set_ylim(0, 1)
axis[1][0].scatter(y_['X'], y_['Y'],)
axis[1][0].set_xlim(0, 0.4)
axis[1][0].set_ylim(0, 1)
axis[1][1].scatter(df['X'], df['Y'], c=model.labels_)
axis[1][1].set_xlim(0, 0.4)
axis[1][1].set_ylim(0, 1)
plt.show()
```
