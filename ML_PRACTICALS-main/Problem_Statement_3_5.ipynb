```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("cities_r2.csv")
```

```python
data.info()
```

```python
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
#For problem statement 3 take column total_graduates instead of effective_literacy_rate_total
selected_col = data[['effective_literacy_rate_total']]
```

```python
sc = StandardScaler()
scaled_col = sc.fit_transform(selected_col)
```

```python
clusters = []
for i in range(1,25):
  kmeans = KMeans(n_clusters=i, n_init=10, random_state=10)
  kmeans.fit(scaled_col)
  clusters.append(kmeans.inertia_)
```

```python
clusters
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,25),clusters, color='blue', linestyle='--',marker='o',markerfacecolor='red',markersize=10)
plt.title("Finding the value of K using Elbow method.")
plt.xlabel("No. of clusters")
plt.ylabel("Clsuter_Score")
plt.show()
```

```python
#Thus from above figure we get K is eqaul to 4
```

```python
kmeans = KMeans(n_clusters=4,random_state=10)
```

```python
kmeans.fit(scaled_col)
```

```python
pred = kmeans.predict(scaled_col)
```

```python
pred
```

```python
data['cluster']=pred
```

```python
data.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(scaled_col,data['population_total'],c=data['cluster'],marker='o',s=50)
plt.title("Clusters craeeted by KMeans Algorithm")
plt.xlabel("Total Graduates")
```
