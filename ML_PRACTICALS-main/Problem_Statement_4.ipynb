```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("cities_r2.csv")
```

```python
data.info()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
xx = data[['effective_literacy_rate_total']]
```

```python
sc = StandardScaler()
x = sc.fit_transform(xx)
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(8,6))
plt.title("Dendrogram")
plt.xlabel("cities")
plt.ylabel("distance")
dend = shc.dendrogram(shc.linkage(x,method='ward'))
```

```python
#From above figure we can determine the no. of cludters = 3
```

```python
cluster = AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage='ward')
pred=cluster.fit_predict(x)
```

```python
pred
```

```python
data['cluster']=pred
```

```python
data.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(xx,xx.index,c=data['cluster'],cmap='rainbow')
plt.title("Agglomerative Clustering")
plt.xlabel('Cities')
plt.ylabel('Effective Literacy Rate')
plt.show()
```
