```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("Hitters.csv")
```

```python
data.shape
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
null_col = ['Salary']
```

```python
from sklearn.impute import SimpleImputer
```

```python
null_imputer = SimpleImputer(strategy="mean")
```

```python
data[null_col] = null_imputer.fit_transform(data[null_col])
```

```python
data.isna().sum()
```

```python
x = data[['CRuns']]
```

```python
sc = StandardScaler()
x_scaled=sc.fit_transform(x)
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(8,6))
plt.title("Dendrogram")
plt.xlabel("CRuns")
plt.ylabel("Distance")
dend = shc.dendrogram(shc.linkage(x,method='ward'))
```

```python
# By above figure we can detrmine the no. odf clusters required are 3
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
pred= cluster.fit_predict(x)
```

```python
pred
```

```python
data['cluster']=pred
```

```python
data.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x,x.index,c=data['cluster'],cmap='rainbow')
plt.title("Agglomeartive Clustering")
plt.xlabel("CRuns")
plt.ylabel("Index")
plt.show()
```
