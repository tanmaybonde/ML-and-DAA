```python
import pandas as pd
import numpy as np
from apyori import apriori
#from mlxtend.frequent_patterns import association_rules
```

```python
df=pd.read_csv('Order2.csv')
df
```

```python
dt=[]
matrix = df.to_numpy()
for i in range(0,7500):
    row = []
    for j in range(0, 20):
        if str(matrix[i][j]) != 'nan':
            row.append(matrix[i][j])
    dt.append(row)
```

```python
len(dt)
```

```python
ass=apriori(dt,min_support=0.03,min_confidence=0.04)
ass_n=list(ass)
```

```python
print(len(ass_n))
```
