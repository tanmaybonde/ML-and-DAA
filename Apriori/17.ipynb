```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from apyori import apriori
```

```python
df = pd.read_csv('Oder3.csv')
```

```python
df.head(10)
```

```python
df.isna().sum()
```

```python
transactions = []
row = []
matrix = df.to_numpy()
row.append(matrix[0][1])
for i in range(1, df.shape[0]):
    if(matrix[i][0] != matrix[i-1][0]):
        transactions.append(row)
        row = []
        row.append(matrix[i][1])
    else:
        row.append(matrix[i][1])
```

```python
transactions
```

```python
associations = apriori(transactions, min_support = 0.003, min_confidence = 0.003)
associations = list(associations)
```

```python
print(len(associations))
```

```python
for x in associations:
    print(x, '\n')
```
